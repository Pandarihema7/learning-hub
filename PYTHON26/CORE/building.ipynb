```python
class A:
    def __init__ (self,q,r):
        self.q=q
        self.r=r
    @classmethod
    def divideby10(cls,n):
        return cls(n//10,n%10)
obj=A.divideby10(25)
print(obj.q)
print(obj.r)
```

```python
class building:
    total_rooms=10
    def __init__ (self,name,road_no):
        self.name=name
        self.road_no=road_no
    @classmethod
    def change_rooms(cls,trooms):
        cls.total_rooms=trooms
b1=building("mb",1)
b2=building("pb",2)
b1.change_rooms(20)
building.change_rooms(15)
print(b1.total_rooms)
print(b2.total_rooms)
```
